```python
###################################### DESKEW BLOCK #############################################

def deskew(FrameNumber):
    
    # get the image
    loadname = 'SinglePages/' + str(FrameNumber) + '.jpg'
    img = cv2.imread(loadname,0)
    img = img[100:-100,100:-100]

    #thresholding the image to a binary image
    thresh,img_bin = cv2.threshold(img,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    #inverting the image
    img_bin = 255-img_bin

    # countcol(width) of kernel as 100th of total width
    kernel_len = np.array(img).shape[1]//100

    # Defining a horizontal kernel to detect all horizontal lines of image
    hor_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_len, 1))

    #Use horizontal kernel to detect and save the horizontal lines in a jpg
    image_2 = cv2.erode(img_bin, hor_kernel, iterations=3)
    horizontal_lines = cv2.dilate(image_2, hor_kernel, iterations=3)
    
    # Apply edge detection method on the image 
    edges = cv2.Canny(img_bin,50,150,apertureSize = 3)
    
    # Save line detection output
    #edge_name = name + '_edges.jpg'
    #cv2.imwrite(edge_name,edges) 
  
    rho = 1  # distance resolution in pixels of the Hough grid
    theta = np.pi / 360  # angular resolution in radians of the Hough grid
    threshold = 12  # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 600  # minimum number of pixels making up a line
    max_line_gap = 200  # maximum gap in pixels between connectable line segments
    line_image = np.copy(img) * 0  # creating a blank to draw lines on

    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(horizontal_lines, rho, theta, threshold, np.array([]),
                    min_line_length, max_line_gap)
    
    #print lines

    # collect angles
    angles = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            angle = float((y2-y1)/(x2-x1))
            angles.append(angle)
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),5)
             
    # Save it
    #newname = 'DeskewedPages/' + str(FrameNumber) + '_lines.jpg'
    #cv2.imwrite(newname,line_image)
         
    angle = np.mean(angles)*180/np.pi
    print 'Detected skew to of the image', FrameNumber, 'to be', angle ,'degrees'

    # rotate the image to deskew it
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    img_deskew = cv2.warpAffine(img, M, (w, h),
                         flags=cv2.INTER_CUBIC, 
                         borderMode=cv2.BORDER_REPLICATE)

    # Save it
    newname = 'DeskewedPages/' + str(FrameNumber) + '.jpg'
    cv2.imwrite(newname,img_deskew)
    
################################### DESKEW WRAPPER ########################################### 
for i in range(1,181):
    deskew(i)
```

```python
################################### TABLE EXTRACTION ###########################################


def GetTables(contours,PageNumber,img):
    # creating a blank to draw lines on
    cell_image1 = np.copy(img)
    #cell_image2 = np.copy(img)
    # Filter contours
    c_filtered = []
    sol_filtered = []
    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if h>80 and w<1000:
            c_filtered.append([x,y,w,h])
        if h<80 and h>40 and w>140 and w<170:
            sol_filtered.append([y,h])
            #cv2.rectangle(cell_image2,(x,y),(x+w,y+h),(0,255,0),2)
    #Get tables
    tables = []
    table = []
    sorted_c = sorted(c_filtered, key = lambda x:x[0])
    x_min = sorted_c[0][0]
    x_max = sorted_c[-1][0] + sorted_c[-1][2]
    width = x_max - x_min
    width = width / 10
    y_scan = 0
    for i in c_filtered:
        if (i[1] - y_scan) > 5:
            if table != [] and len(table) == 10:
                # Find the solution row
                y_exp = y_scan + h_scan
                h_exp = h_scan
                Flag = True
                y_thrashold = 20
                while Flag:
                    Flag = False
                    for j in sol_filtered:
                        if abs(y_exp - j[0]) < y_thrashold:
                            y_exp = j[0] + j[1]
                            h_exp = j[1]
                            Flag = True
                table.append([y_exp-h_exp,h_exp])
                #cv2.rectangle(cell_image2,(x_min,y_exp-h_exp),(x_max,y_exp),(0,255,0),2)
                tables.append(table)
                #for j in table:
                for j in table[:-1]:
                    x = j[0]
                    y = j[1]
                    w = j[2]
                    h = j[3]
                    cv2.rectangle(cell_image1,(x,y),(x+w,y+h),(0,255,0),2)
            table = [i]
            y_scan = i[1]
            h_scan = i[3]
        elif (i[2] / width) > 1.5:
            if abs(i[3] - h_scan) > 10:
                print 'TROUBLE WITH CELL WIDTH AND HEIGHT DETECTION'
            cutNumber = int(round(i[2] / width))
            w_new = int(i[2] / cutNumber)
            for j in range(cutNumber):
                x_new = i[0] + w_new * j
                table.append([x_new,i[1],w_new,i[2]])
        else:
            if abs(i[3] - h_scan) > 10:
                print 'TROUBLE WITH CELL HEIGHT DETECTION'
            table.append(i)
            

    if table != [] and len(table) == 10:
        # Find the solution row
        y_exp = y_scan + h_scan
        h_exp = h_scan
        Flag = True
        y_thrashold = 20
        while Flag:
            Flag = False
            for j in sol_filtered:
                if abs(y_exp - j[0]) < y_thrashold:
                    y_exp = j[0] + j[1]
                    h_exp = j[1]
                    Flag = True
        table.append([y_exp-h_exp,h_exp])
        #cv2.rectangle(cell_image2,(x_min,y_exp-h_exp),(x_max,y_exp),(0,255,0),2)
        tables.append(table)
        #for j in table:
        #for j in table[:-1]:
            #x = j[0]
            #y = j[1]
            #w = j[2]
            #h = j[3]
            #cv2.rectangle(cell_image1,(x,y),(x+w,y+h),(0,255,0),2)
    
    # get the image
    cellname = 'ExtrTest/' + str(PageNumber) + '_cells1.jpg'
    cv2.imwrite(cellname, cell_image1)
    #cellname = 'ExtractedTables/' + str(PageNumber) + '_cells2.jpg'
    #cv2.imwrite(cellname, cell_image2)
    
    return tables

# Sorting of countours and generation of bounding boxes
def sort_contours(cnts, method="left-to-right"):
    # initialize the reverse flag and sort index
    reverse = False
    i = 0
    # handle if we need to sort in reverse
    if method == "right-to-left" or method == "bottom-to-top":
        reverse = True
    # handle if we are sorting against the y-coordinate rather than
    # the x-coordinate of the bounding box
    if method == "top-to-bottom" or method == "bottom-to-top":
        i = 1
    # construct the list of bounding boxes and sort them from top to
    # bottom
    boundingBoxes = [cv2.boundingRect(c) for c in cnts]
    (cnts, boundingBoxes) = zip(*sorted(zip(cnts, boundingBoxes),
    key=lambda b:b[1][i], reverse=reverse))
    # return the list of sorted contours and bounding boxes
    return (cnts, boundingBoxes)

def ConvertNonASCII(TesOut):
    TesStr = filter(lambda x: x in printable, TesOut)  
    if TesStr == None:
        TesStr = '?'
    return TesStr

def GetNumberCoord(stripe):
    x_mean = [np.mean(stripe[i,:]) for i in range(stripe.shape[0])]
    Number = False
    events = []
    Threshold = 252
    for i in range(stripe.shape[0]):
        if not Number and x_mean[i]<Threshold:
            Number = True
            events.append(i)
        if Number and x_mean[i] > Threshold:
            Number = False
            events.append(i)
    # get coordinates of number edges
    edges = []
    for i in range(int(len(events) / 2)):
        height = events[2*i + 1] - events[2*i] 
        if height > 24 and height < 32:
            edges.append([events[2*i],events[2*i+1]])
    return edges

def ExtractTables(PageNumber):

    # get the image
    loadname='DeskewedPages/'+str(PageNumber)+'.jpg'
    img = cv2.imread(loadname,0)

    # thresholding the image to a binary image
    thresh,img_bin = cv2.threshold(img,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # inverting the image
    img_bin = 255-img_bin

    # countcol(width) of kernel as 100th of total width
    kernel_len = np.array(img).shape[1]//100

    # Defining a vertical kernel to detect all vertical lines of image
    ver_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, kernel_len))

    # Defining a horizontal kernel to detect all horizontal lines of image
    hor_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_len, 1))

    # A kernel of 2x2
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
    
    # Use vertical kernel to detect and save the vertical lines in a jpg
    image_1 = cv2.erode(img_bin, ver_kernel, iterations=3)
    vertical_lines = cv2.dilate(image_1, ver_kernel, iterations=3)

    # Use horizontal kernel to detect and save the horizontal lines in a jpg
    image_2 = cv2.erode(img_bin, hor_kernel, iterations=3)
    horizontal_lines = cv2.dilate(image_2, hor_kernel, iterations=3)

    # Combine horizontal and vertical lines in a new third image, with both having same weight.
    img_vh = cv2.addWeighted(vertical_lines, 0.5, horizontal_lines, 0.5, 0.0)

    # Eroding and thesholding the image
    img_vh = cv2.erode(~img_vh, kernel, iterations=2)
    thresh, img_vh = cv2.threshold(img_vh,128,255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    bitxor = cv2.bitwise_xor(img,img_vh)
    bitnot = cv2.bitwise_not(bitxor)
    
    # get the image
    cellname = 'ExtrTest/' + str(PageNumber) + '_vh.jpg'
    cv2.imwrite(cellname, img_vh)

    # Detect contours for following box detection
    contours, hierarchy = cv2.findContours(img_vh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Sort all the contours by top to bottom.
    contours, boundingBoxes = sort_contours(contours, method="top-to-bottom")

    tables = GetTables(contours,PageNumber,img)

    # get the image
    #cellname = 'ExtractedTables/' + str(PageNumber) + '_cells.jpg'
    #cv2.imwrite(cellname, cell_image)

    ######################### EXTRACT TABLES (10 SUMS EACH) #################################
        
    iterator = 0
    page = []
    for i in range(len(tables)):
        table = tables[i][:10]
        if len(table) != 10:
            print '10 cells did not found'
        
        # Sort contours
        table_sorted = []
        TableX = [k[0] for k in table]
        Index = np.argsort(TableX)
        for j in Index:
            table_sorted.append(table[j])
            
        # update table
        table = table_sorted
            
        # Get the dimension of the table
        x1 = 10000
        x2 = 0
        y1 = 10000
        y2 = 0
        y1min = 0
        for cell in table:
            if cell[0]<x1:
                x1=cell[0]
            if cell[1]<y1:
                y1=cell[1]
            if cell[1]>y1min:
                y1min=cell[1]
            if cell[2]+cell[0]>x2:
                x2=cell[2]+cell[0]
            if cell[1]+cell[3]>y2:
                y2=cell[1]+cell[3]
                
        # Solution frame
        y_sol1 = tables[i][10][0]
        y_sol2 = y_sol1 + tables[i][10][1]
        Sframe = img[y_sol1:y_sol2,x1:x2]
    
        # Table frame
        Tframe = img[y1:y2,x1:x2]
        
        # Combine frames
        Tframe = np.vstack((Tframe,Sframe))
        
        #fileName = 'ExtractedTables/' + str(PageNumber) + '.' + str(i) +'_frame.jpg'
        #cv2.imwrite(fileName,Tframe)
        
        ############################### DESKEW THE TABLE ######################
        # Get the skew_angle
        skew_angle = float((table[9][1]-table[0][1])/(table[9][0]-table[0][0]))*180/np.pi

        # rotate the table to deskew it
        (h, w) = Tframe.shape[:2]
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, skew_angle, 1.0)
        Tframe_deskew = cv2.warpAffine(Tframe, M, (w, h),
                         flags=cv2.INTER_CUBIC, 
                         borderMode=cv2.BORDER_REPLICATE)
        
        # Save it
        #fileName = 'ExtractedTables/' + str(PageNumber) + '.' + str(i) +'_frame.jpg'
        #cv2.imwrite(fileName,Tframe_deskew)
        
        
        #################### REMOVE VERTICAL SEPARATION LINES ###################
        
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 1))
        thresh,Tframe_bin = cv2.threshold(Tframe_deskew,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
        
        # Remove separation lines
        CutStripe = 20
        DetectionBorder = 8
        tableScript = []
        for stripe in table:
            primer = []
            l = int(stripe[0] + CutStripe - x1)
            r = int(stripe[0] + stripe[2] - CutStripe - x1)
            stripeFrame = Tframe_bin[:,l:r]
            NumbC = GetNumberCoord(stripeFrame)
            for coord in NumbC:
                t = coord[0] - DetectionBorder
                b = coord[1] + DetectionBorder
                NumberFrame = stripeFrame[t:b,:]
                border = cv2.copyMakeBorder(NumberFrame,2,2,2,2, cv2.BORDER_CONSTANT,value=[255,255])
                dilation = cv2.dilate(border, kernel,iterations=1)
                erosion = cv2.erode(dilation, kernel,iterations=2)
                out = pytesseract.image_to_string(erosion, config='--psm 6')
                out = ConvertNonASCII(out)
                primer.append(out)
            tableScript.append(primer)
        page.append(tableScript)
        fileName = 'ExtrTest/' + str(PageNumber) + '.' + str(i+1) +'.txt'
        print PageNumber, '.', str(i+1)
        with open(fileName, 'w') as file:
            writer = csv.writer(file)
            writer.writerows(tableScript)
            
            
            
################################### TABLE EXTRACTION WRAPPER ###########################################          

#ExtractTables(65) 
```
